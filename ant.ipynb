```python
import math
import random
import time

import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
import mlx.core as mx
import scipy
```

```python
NROW = 5
NCOL = 5
WHITE = 0
BLACK = 1
ANT = 'X'

def gridnew(nrow, ncol):
    return [[0 for _ in range(ncol)] for _ in range(nrow)]

def turn(dir):
    if dir == 'N':
        return 'E'
    elif dir == 'E':
        return 'S'
    elif dir == 'S':
        return 'W'
    elif dir == 'W':
        return 'N'
    else:
        assert False, "simuate: unsupported direction"

def flip(grid, pos):
    (row, col) = pos
    if grid[row][col] == WHITE:
        grid[row][col] = BLACK
    elif grid[row][col] == BLACK:
        grid[row][col] = WHITE
    else:
        assert False, "flip: unsupported color"

def step(pos, dir):
    (row, col) = pos
    if dir == 'N':
        return row - 1, col
    elif dir == 'E':
        return row, col + 1
    elif dir == 'S':
        return row + 1, col
    elif dir == 'W':
        return row, col - 1
    
def gridprint(grid, antpos):
    nrow = len(grid)
    ncol = len(grid[0])
    for row in range(nrow):
        for col in range(ncol - 1):
            if (row, col) == antpos:
                print(f'{ANT} ', end='')
            else:
                print(f"{grid[row][col]} ", end='')
        if (row, ncol - 1) == antpos:
            print(f"{ANT}")
        else:
            print(f"{grid[row][ncol - 1]}")
            

# TODO: this mutates the grid
# TODO: handle edge behavior
# TODO: I think this orientation is off?
def simulate(grid, pos=(0, 0), dir='E', nstep=10, interval=1):
    row, col = pos if pos is not None else (len(grid) // 2, len(grid[0]) // 2)

    gridprint(grid, (row, col))
    print('---' * len(grid[0]))

    for _ in range(nstep):
        if grid[row][col] == WHITE:
            dir = turn(dir)
            flip(grid, (row, col)) 
            row, col = step((row, col), dir)
        elif grid[row][col] == BLACK:
            dir = turn(turn(turn(dir)))
            flip(grid, (row, col))
            row, col = step((row, col), dir) 
        else:
            assert False, "simulate: unsupported color"
        print()
        gridprint(grid, (row, col)) 
        if interval > 0:
            time.sleep(interval)


grid = gridnew(11, 11)
simulate(grid, pos=None)
```

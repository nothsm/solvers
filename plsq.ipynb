```python
import math
import random

import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
import mlx.core as mx
import mpmath
import scipy
```

```python
K, k = sp.symbols('K k')

bbp = sp.Sum((1 / 16**k) * ((4 / (8 * k + 1)) - (2 / (8 * k + 4)) - (1 / (8 * k + 5)) - (1 / (8 * k + 6))), (k, 0, K))

print("K | S_K | eps")
for i in range(5):
    val = bbp.subs('K', i).evalf()
    print(f"{i} | {val} | {abs(val - sp.pi).evalf()}")
```

```python
log2expr = sp.Sum(1 / (k * (2 ** k)), (k, 1, K))
target = sp.log(2)

print(target.evalf())
print("K | S_K | eps")
print('---------------')
for i in range(64):
    val = log2expr.subs('K', i).evalf()
    print(f"{i} | {val} | {abs(val - sp.log(2)).evalf()}")
```

```python
sign, mantissa, exponent, nbits = target.evalf()._mpf_

print(target.evalf())

print()
print(sign)
print(mantissa)
print(bin(mantissa))
print(nbits)
```

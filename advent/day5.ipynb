```python
def day5read(filename):
    intervals = []
    ingredients = []
    with open(filename, 'r') as f:
        # read intervals
        while (line := f.readline()) != '\n':
            beg, end = line.strip().split('-')
            intervals.append((int(beg), int(end) + 1))
        # read ingredients
        while (line := f.readline()) != '':
            ingredients.append(int(line.strip()))

    # merge intervals
    # TODO: Check off-by-one errors
    intervals = sorted(intervals, key=lambda interval: interval[0])
    merged = []
    for (beg, end) in intervals:
        if len(merged) == 0:
            merged.append((beg, end))
        else:
            tmp_beg, tmp_end = merged[-1] 
            if beg >= tmp_end:
                merged.append((beg, end))
            else:
                merged.pop()
                merged.append((min(beg, tmp_beg), max(end, tmp_end)))
    return merged, ingredients

def solve1(intervals, ingredients):
    nfresh = 0
    for i in ingredients:
        nfresh += int(any(beg <= i < end for (beg, end) in intervals))
    return nfresh

def solve2(intervals):
    return sum(end - beg for (beg, end) in intervals)


filename = 'data/day5.txt'
intervals, ingredients = day5read(filename)
print(solve1(intervals, ingredients))
print(solve2(intervals))
```

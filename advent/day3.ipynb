```python
NBAT_P1 = 2
NBAT_P2 = 12

# TODO: Precompute postfix max
def part1(banks):
    bats = []
    for bank in banks:
        i0 = max(range(len(bank) - (NBAT_P1 - 1)), key=lambda i: int(bank[i]))
        i1 = max(range(i0 + 1, len(bank) - (NBAT_P1 - 2)), key=lambda i: int(bank[i]))
        bats.append(int(bank[i0] + bank[i1]))
    return sum(bats)

def part2(banks):
    bats = []
    for bank in banks:
        ixs = []
        bs = []

        ix0 = max(range(len(bank) - (NBAT_P2 - 1)), key=lambda i: int(bank[i]))
        ixs.append(ix0)
        bs.append(bank[ix0])

        for k in range(2, NBAT_P2 + 1):
            ix = max(range(ixs[-1] + 1, len(bank) - (NBAT_P2 - k)), key=lambda i: int(bank[i]))
            ixs.append(ix)
            bs.append(bank[ix])
        bats.append(int(''.join(bs)))
    return sum(bats)

def day3read(filename):
    banks = []
    with open(filename, 'r') as f:
        for line in f:
            banks.append(line.strip())
    return banks 

def solve1(filename):
    banks = day3read(filename)
    jolt = part1(banks)
    return jolt 

def solve2(filename):
    banks = day3read(filename)
    jolt = part2(banks)
    return jolt

filename = 'data/day3.txt'
solve1(filename)
solve2(filename)
```

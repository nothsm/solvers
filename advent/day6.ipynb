```python
MUL = '*'
ADD = '+'

def day6read(filename):
    lines = []
    with open(filename, 'r') as f:
        for line in f:
            words = line.strip().split(' ')
            lines.append([w for w in words if len(w) > 0])
    for i in range(len(lines) - 1):
        lines[i] = [int(w) for w in lines[i]]

    assert all(len(lines[i]) == len(lines[0]) for i in range(len(lines)))

    problems = []
    for col in range(len(lines[0])):
        problem = []
        for row in range(len(lines) - 1):
            problem.append(lines[row][col])
        op = lines[len(lines) - 1][col]
        problems.append((op, problem))

    return problems

def day6read2(filename):
    lines = []
    with open(filename, 'r') as f:
        lines = [line for line in f.readlines()]

    nrow = len(lines)
    ncol = len(lines[-1])

    # transpose lines
    transposed = [[None] * nrow for _ in range(ncol)]
    for row in range(nrow):
        for col in range(ncol):
            transposed[col][row] = lines[row][col]
    transposed = [''.join(words).strip() for words in transposed]

    problems = []
    i = 0
    while i < len(transposed):
        j = i
        op = ''
        nums = []
        # read a single problem
        while j < len(transposed) and len((line := transposed[j])) > 0:
            if line[-1] == ADD:
                assert op == ''
                num, op = line[:-1], line[-1]
                nums.append(int(num))
            elif line[-1] == MUL:
                assert op == ''
                num, op = line[:-1], line[-1]
                nums.append(int(num))
            else:
                nums.append(int(line))
            j += 1
        problems.append((op, list(reversed(nums))))
        i = j + 1
    return list(reversed(problems)) # none of the reversing is necessary. it's just to match the spec

def solve1(problems):
    totals = []
    for (op, nums) in problems:
        if op == ADD:
            total = 0
            for num in nums:
                total += num 
            totals.append(total)
        elif op == MUL:
            total = 1
            for num in nums:
                total *= num
            totals.append(total)
        else:
            raise NotImplementedError("operation not supported")
    return sum(totals)

filename = 'data/day6.txt'
problems = day6read(filename)
problems2 = day6read2(filename)
print(solve1(problems))
print(solve1(problems2))
```

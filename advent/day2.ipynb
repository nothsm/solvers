```python
NREPEAT = 2

def day2read(filename):
    with open(filename, 'r') as f:
        line = f.readline()
        words = line.split(',')
        id_strs = [word.split('-') for word in words] 
        return [(int(beg), int(end) + 1) for beg, end in id_strs] 
    
def part1(ids):
    invalids = []
    for (beg, end) in ids:
        invalid = []
        for i in range(beg, end):
            i_str = str(i)

            if not (len(i_str) % NREPEAT == 0):
                continue

            k = len(i_str) // NREPEAT

            assert NREPEAT * k == len(i_str)

            if NREPEAT * i_str[:k] == i_str:
                invalid.append(i)

        invalids.append(invalid)
    return sum(sum(invalid) for invalid in invalids) 

def part2(ids):
    invalids = []
    for (beg, end) in ids:
        invalid = []
        for i in range(beg, end):
            i_str = str(i)

            for nrepeat in range(NREPEAT, len(i_str) + 1):
                if not (len(i_str) % nrepeat == 0):
                    continue

                k = len(i_str) // nrepeat 

                assert nrepeat * k == len(i_str)

                if nrepeat * i_str[:k] == i_str:
                    invalid.append(i)
                    break

        invalids.append(invalid)
    return sum(sum(invalid) for invalid in invalids) 

def solve1(filename):
    ids = day2read(filename)
    return part1(ids)

def solve2(filename):
    ids = day2read(filename)
    return part2(ids)

filename = 'data/day2.txt'
print(solve1(filename))
print(solve2(filename))
```

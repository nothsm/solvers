```python
import heapq
from collections import defaultdict

K_LARGEST = 3
N_STEPS = 1000

def day8read(filename):
    boxes = []
    with open(filename, 'r') as f:
        for line in f:
            x, y, z = line.strip().split(',')
            boxes.append((int(x), int(y), int(z)))
    return boxes

def l2dist(x, y):
    x1, x2, x3 = x
    y1, y2, y3 = y
    return (((x1 - y1) ** 2) + ((x2 - y2) ** 2) + ((x3 - y3) ** 2)) ** 0.5

def solve1(boxes):
    assert len(boxes) == len(set(boxes))

    # TODO: Is there a more efficient way of implementing this?
    heap = []
    for i, b0 in enumerate(boxes):
        for b1 in boxes[i + 1:]:
            assert b0 != b1
            heapq.heappush(heap, (l2dist(b0, b1), b0, b1))

    # I can store the indices of the boxes rather than the boxes themselves

    circ_id = 0
    box2circ = {}
    circ2box = defaultdict(set)
    for _ in range(N_STEPS):
        dist, b0, b1 = heapq.heappop(heap)

        if b0 in box2circ and b1 in box2circ:
            circ0, circ1 = box2circ[b0], box2circ[b1]

            bs = circ2box.pop(circ1)
            for b in bs:
                box2circ[b] = circ0
                circ2box[circ0].add(b)

            box2circ[b1] = circ0
            circ2box[circ0].add(b1)
        elif b0 in box2circ and b1 not in box2circ:
            circ = box2circ[b0]
            box2circ[b1] = circ
            circ2box[circ].add(b1)
        elif b0 not in box2circ and b1 in box2circ:
            circ = box2circ[b1]
            box2circ[b0] = circ
            circ2box[circ].add(b0)
        elif b0 not in box2circ and b1 not in box2circ:
            circ = circ_id
            circ_id += 1
            box2circ[b0] = circ
            box2circ[b1] = circ
            circ2box[circ].add(b0)
            circ2box[circ].add(b1)

    sizeprod = 1
    for circ in sorted(circ2box.values(), key=len, reverse=True)[:K_LARGEST]:
        sizeprod *= len(circ)
    return sizeprod 

def solve2(boxes):
    assert len(boxes) == len(set(boxes))

    # TODO: Is there a more efficient way of implementing this?
    heap = []
    for i, b0 in enumerate(boxes):
        for b1 in boxes[i + 1:]:
            assert b0 != b1
            heapq.heappush(heap, (l2dist(b0, b1), b0, b1))

    # I can store the indices of the boxes rather than the boxes themselves

    circ_id = 0
    box2circ = {}
    circ2box = defaultdict(set)

    last_b0 = None
    last_b1 = None
    while heap:
        dist, b0, b1 = heapq.heappop(heap)

        if b0 in box2circ and b1 in box2circ:
            circ0, circ1 = box2circ[b0], box2circ[b1]

            bs = circ2box.pop(circ1)
            for b in bs:
                box2circ[b] = circ0
                circ2box[circ0].add(b)

            box2circ[b1] = circ0
            circ2box[circ0].add(b1)

            if len(circ2box[circ0]) == len(boxes):
                last_b0, last_b1 = b0, b1
                break

        elif b0 in box2circ and b1 not in box2circ:
            circ = box2circ[b0]
            box2circ[b1] = circ
            circ2box[circ].add(b1)

            if len(circ2box[circ]) == len(boxes):
                last_b0, last_b1 = b0, b1
                break

        elif b0 not in box2circ and b1 in box2circ:
            circ = box2circ[b1]
            box2circ[b0] = circ
            circ2box[circ].add(b0)

            if len(circ2box[circ]) == len(boxes):
                last_b0, last_b1 = b0, b1
                break

        elif b0 not in box2circ and b1 not in box2circ:
            circ = circ_id
            circ_id += 1
            box2circ[b0] = circ
            box2circ[b1] = circ
            circ2box[circ].add(b0)
            circ2box[circ].add(b1)


            if len(circ2box[circ]) == len(boxes):
                last_b0, last_b1 = b0, b1
                break
    x0, _, _ = last_b0
    x1, _, _ = last_b1
    return x0 * x1

filename = 'data/day8.txt'
boxes = day8read(filename)
print(solve1(boxes))
print(solve2(boxes))
```

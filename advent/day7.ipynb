```python
START = 'S'
SPLITTER = '^'
EMPTY = '.'
BEAM = '|'

def day7read(filename):
    with open(filename, 'r') as f:
        return [line.strip() for line in f.readlines()]
    
def solve1(grid):
    timeline = [{grid[0].index(START)}] # TODO: Can pre-allocate for higher performance
    nsplit = 0
    for row in range(1, len(grid) - 1):
        bs = set()
        for col in timeline[-1]:
            if grid[row + 1][col] == EMPTY:
                bs.add(col)
            elif grid[row + 1][col] == SPLITTER:
                bs.add(col - 1)
                bs.add(col + 1)
                nsplit += 1
            else:
                raise NotImplementedError()
        timeline.append(bs)
    return nsplit, timeline 

def solve2(grid):
    nrow = len(grid)
    ncol = len(grid[0])

    cache = {}
    def backtrack(row, col):
        if row == nrow:
            return 1
        else:
            if (row, col) in cache:
                return cache[(row, col)]
            elif grid[row][col] == EMPTY:
                ret = backtrack(row + 1, col)
                cache[(row, col)] = ret
                return ret
            elif grid[row][col] == SPLITTER:
                ret = backtrack(row + 1, col - 1) + backtrack(row + 1, col + 1)
                cache[(row, col)] = ret
                return ret
            else:
                raise NotImplementedError()
    return backtrack(1, grid[0].index(START))
    
filename = 'data/day7.txt'
grid = day7read(filename)
nsplit, _ = solve1(grid)
print(nsplit)
print(solve2(grid))
```

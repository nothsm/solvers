```python
DIAL_BEG = 50
DIAL_LEN = 100
DIAL_TGT = 0
LEFT = 'L'
RIGHT = 'R'

def part1(ts):
    dial = DIAL_BEG
    nzero = 0
    for t in ts:
        dial = (dial + t) % DIAL_LEN
        nzero += + int(dial == DIAL_TGT)
    return dial, nzero

# TODO: Is there a more efficient solution for this?
def part2(ts):
    dial = DIAL_BEG 
    nzero = 0
    for t in ts:
        if t >= 0:
            for _ in range(t):
                dial = (dial + 1) % DIAL_LEN 
                nzero = nzero + int(dial == DIAL_TGT)
        elif t < 0:
            for _ in range(abs(t)):
                dial = (dial - 1) % DIAL_LEN
                nzero = nzero + int(dial == DIAL_TGT)
    return dial, nzero

# read and parse file
def day1read(filename):
    ts = []
    with open(filename, 'r') as f:
        for line in f:
            if line[0] == LEFT:
                _, num = line.split(LEFT)
                ts.append(-int(num))
            elif line[0] == RIGHT:
                _, num = line.split(RIGHT)
                ts.append(int(num))
            else:
                assert False
    return ts

def solve1(filename):
    ts = day1read(filename)
    _, nzero = part1(ts)
    return nzero

def solve2(filename):
    ts = day1read(filename)
    _, nzero = part2(ts)
    return nzero

data_file = 'data/day1.txt'
print(solve1(data_file))
print(solve2(data_file))
```

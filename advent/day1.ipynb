```python
import os
```

```python
LEFT = 'L'
RIGHT = 'R'

def dialeval(turns, beg=50, dial_len=100):
    history = [beg]
    for turn in turns:
        cur = history[-1]
        history.append((cur + turn) % dial_len)
    return history 


def solve1(filename):
    # parse file
    turns = []
    with open(filename, 'r') as f:
        for line in f:
            if line[0] == LEFT:
                _, num = line.split(LEFT)
                turns.append(-int(num))
            elif line[0] == RIGHT:
                _, num = line.split(RIGHT)
                turns.append(int(num))
            else:
                assert False
    history = dialeval(turns, beg=50, dial_len=100)
    return sum(1 for state in history if state == 0) 

data_file = 'data/day1.txt'
solve1(data_file)
```

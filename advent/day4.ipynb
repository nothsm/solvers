```python
EMPTY = '.'
ROLL = '@'
FORKLIFT = 'x'
ROLLCAP = 4

def part1(grid):
    nrow = len(grid)
    ncol = len(grid[0])

    accessible = []
    for row in range(nrow):
        for col in range(ncol):
            if grid[row][col] == ROLL:
                # get adjacent rolls
                rolls = []
                for r in range(row - 1, row + 2):
                    for c in range(col - 1, col + 2):
                        if 0 <= r < nrow and 0 <= c < ncol and (r, c) != (row, col) and grid[r][c] == ROLL:
                            rolls.append((r, c))
                # a roll can be accessed by forklift if fewer than ROLLCAP adjacent rolls
                if len(rolls) < ROLLCAP:
                    accessible.append((row, col))
    return accessible, len(accessible)

def part2(grid):
    removed = []
    nremove = 0
    while True:
        accessible, nroll = part1(grid)
        for (r, c) in accessible:
            removed.append((r, c))
            grid[r][c] = EMPTY
        nremove += nroll
        if nroll == 0:
            break
    return removed, nremove

# TODO: Can use a dict, too
def day4read(filename):
    grid = []
    with open(filename, 'r') as f:
        for line in f:
            grid.append(list(line.strip()))
    return grid 

def solve1(filename):
    grid = day4read(filename)
    _, nroll = part1(grid)
    return nroll

def solve2(filename):
    grid = day4read(filename)
    _, nremove = part2(grid)
    return nremove

filename = 'data/day4.txt'
print(solve1(filename))
print(solve2(filename))
```

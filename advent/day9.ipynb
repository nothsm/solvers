```python
def day9read(filename):
    tiles = []
    with open(filename, 'r') as f:
        for line in f:
            x, y = line.strip().split(',')
            tiles.append((int(x), int(y)))
    return tiles

def rect(t0, t1):
    x0, y0 = t0
    x1, y1 = t1
    return (abs(x1 - x0) + 1) * (abs(y1 - y0) + 1)

def solve1(tiles):
    return max(rect(t0, t1) for i, t0 in enumerate(tiles) for t1 in tiles[i + 1:])

filename = 'data/day9.txt'
tiles = day9read(filename)
solve1(tiles)
```

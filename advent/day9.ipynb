```python
from itertools import pairwise

def rectarea(t0, t1):
    i0, j0 = t0
    i1, j1 = t1

    length = abs(j1 - j0) + 1
    width = abs(i1 - i0) + 1
    return length * width 

def rect(t0, t1):
    i0, j0 = t0
    i1, j1 = t1

    i_min, j_min = min(i0, i1), min(j0, j1)
    i_max, j_max = max(i0, i1), max(j0, j1)

    R = {(i, j) for i in range(i_min, i_max + 1) for j in range(j_min, j_max + 1)}
    assert len(R) == rectarea(t0, t1)
    return R

def day9read(filename):
    tiles = []
    with open(filename, 'r') as f:
        for line in f:
            col, row = line.strip().split(',')
            tiles.append((int(row), int(col)))
    return tiles

def is_interior(tile, perimeter):
    i_min, j_min = 0, 0
    i_max, j_max = max(i for i, j in perimeter), max(j for i, j in perimeter)

    i_t, j_t = tile
    while True:
        if i_t <= i_min or i_t >= i_max:
            ...


# what if I just save the rectangles/tiles that are no longer allowed?
# idea: from the target point, walk along the points until something is inside the rectangle OR ...
#       - adjacent points are easy
#       - can probably use the containment condition to further improve the algorithm

def solve1(tiles):
    return max(rectarea(t0, t1) for i, t0 in enumerate(tiles) for t1 in tiles[i + 1:])

def solve2(tiles):
    green = set()
    for t0, t1 in pairwise(tiles):
        green = green | rect(t0, t1)

    # TODO: remove red from green

    seen = set()
    def dfs(i, j):
        if (i, j) in seen or i < 0 or j < 0 or (i, j) in tiles:
            return
        else:
            green.add((i, j))
            seen.add((i, j))
            dfs(i + 1, j)
            dfs(i - 1, j)
            dfs(i, j + 1)
            dfs(i, j - 1)



    # dfs(*next(iter(green)))
    return sorted(green)

filename = 'data/day9-small.txt'
tiles = day9read(filename)
print(solve1(tiles))
print(solve2(tiles))
```

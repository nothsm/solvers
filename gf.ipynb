```python
def a(n):
    if n == 0:
        return 0
    else:
        return 2 * a(n - 1) + 1
    
[a(n) for n in range(11)]
```

```python
x = sp.symbols('x')

sp.apart(x / ((1 - x) * (1 - 2 * x)))
```

```python
def fib(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fib(n - 1) + fib(n - 2)
    
[fib(n) for n in range(11)]
```

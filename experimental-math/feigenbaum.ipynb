```python
import math
from pprint import pprint

import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
```

```python
def logistic(x):
    return a * x * (1 - x)
```

```python
x = 1 / 3
a = 3.2

n = 50

xs = [x]
for _ in range(n):
    xs.append(logistic(xs[-1]))

print(xs)

fig, ax = plt.subplots()

ax.set_ylim(0.0, 1.0)
ax.set_title(f'Logistic map (x = {x}, a = {a})')

ax.plot(xs, linewidth=2.0, marker='o', markersize=3.0)
```

```python
for a in [2.75, 3.0, 3.25, 3.5, 3.75, 4.0]:
    xs = [x]
    for _ in range(n):
        xs.append(logistic(xs[-1]))

    print(xs)

    fig, ax = plt.subplots(figsize=(6, 4))

    ax.set_ylim(0.0, 1.0)
    ax.set_title(f'Logistic map (x = {x}, a = {a})')

    ax.plot(xs, linewidth=2.0, marker='o', markersize=3.0)
```

```python
num_x = 350
num_drop = 50

for num_a in [10, 50, 100, 200, 300, 400]:
    A = np.linspace(0, 4, num=num_a).tolist()

    xss = []
    for a in A:
        xs = [x]
        for _ in range(num_x):
            xs.append(logistic(xs[-1]))

        xss.append(list(set(xs[num_drop:])))



    fig, ax = plt.subplots()

    ax.set_xlim(0 - 0.05, 4 + 0.05)
    ax.set_ylim(0 - 0.05, 1 + 0.05)

    for (a, xs) in zip(A, xss):
        X = [a for _ in range(len(xs))]
        Y = xs
        ax.scatter(X, Y, s=0.1, c='orange')
```

```python
num_x = 350
num_drop = 50

for num_a in [10, 50, 100, 200, 300, 400]:
    A = np.linspace(0, 1, num=num_a).tolist()

    xss = []
    for a in A:
        xs = [x]
        for _ in range(num_x):
            xs.append(a * math.sin(math.pi * math.sqrt(xs[-1])))

        xss.append(list(set(xs[num_drop:])))



    fig, ax = plt.subplots()

    ax.set_xlim(0 - 0.05, 1 + 0.05)
    ax.set_ylim(0 - 0.05, 1 + 0.05)

    for (a, xs) in zip(A, xss):
        X = [a for _ in range(len(xs))]
        Y = xs
        ax.scatter(X, Y, s=0.1, c='orange')
```

**Example 1.1.** *A proof and a disproof.* Any modern computer algebra system can tell one that
$$
0 < \int_0^1 \frac{(1 - x)^4 x^4}{1 + x^2} dx = \frac{22}{7} - \pi, \quad (1)
$$
since the integral may be interpreted as the area under a positive curve.

If, however, we ask the same system to compute the indefinite integral, we are likely to be told that
$$
\int_0^t \frac{(1 - t)^4 t^4}{1 + t^2} dt = \frac{1}{7} t^7 - \frac{2}{3} t^6 + t^5 - \frac{4}{3} t^3 + 4 t - 4 \arctan(t).
$$
Then (1) is now rigorously established by differentiation and an appeal to the Fundamental Theorem of Calculus.

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
```

```python
x = sp.symbols('x')
t = sp.symbols('t')

sp.init_printing(use_unicode=True)
```

```python
expr = (((1 - x) ** 4) * (x ** 4)) / (1 + (x ** 2))

sp.Integral(expr, (x, 0, 1)).doit()
```

```python
sp.integrate(expr, x)
```
